```python
import pandas as pd
import numpy as np
from multimodel_analysis import MultiModelClassifier

# 1. Load your dataset
# Ensure your data is numeric or pre-encoded
df = pd.read_csv('https://raw.githubusercontent.com/udityamerit/MultiModel-Package-for-Machine-Learning/refs/heads/main/Dataset/diabetes.csv')
X = df.drop('Outcome', axis=1)
y = df['Outcome']
```

```python
model = MultiModelClassifier(X, y, test_size=0.3, scaled_data=True) # it is for calling your classifier model
models = model.run_all_models()
```

```python
model.show_tabular_report(models)
```

```python
model.plot_confusion_matrices(models)
```

```python
model.plot_roc_curves(models)
```

```python
model.get_summary(models)
```

# Regression

```python
df = pd.read_csv('https://raw.githubusercontent.com/udityamerit/Complete-Machine-Learning-For-Beginners/refs/heads/main/0-Dataset/Algerian_forest_fires_dataset_cleaned.csv')
```

```python
df.drop(['day', 'month', 'year'], axis=1, inplace=True)

df['Classes'] = np.where(df['Classes'].str.contains('not fire'), 0, 1) 
```

```python
## Independent and dependent features
X =df.drop('FWI', axis=1)
y = df['FWI']
```

```python
from multimodel_analysis import MultiModelRegressior
reg_model = MultiModelRegressior(X, y, test_size=0.3, scaled_data=True)
mode = reg_model.run_all_models()
```

```python
reg_model.show_tabular_report(mode)
```

```python
reg_model.plot_true_vs_predicted(mode)
```

```python
reg_model.plot_comparison(mode)
```
